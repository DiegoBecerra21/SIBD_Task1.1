
## üîé Step 3: Filtering Rows

Functions to explore:
- df[df["Age"] > 30]

df[df["Age"] > 30] nos permite filtrar filas de nuestro data frame con una condici√≥n l√≥gica.
En el ejemplo de arriba solo seleccionar√° las filas donde la edad sea mayor que 30


## üßπ Step 4: Handling Missing Data

Functions to explore:
- df.isna()
- df.isna().sum()
- df.dropna()
- df.fillna()


df.isna() nos devuelve un data frame del mismo tama√±o que el original donde cada celda muestra:
- True si el valor est√° vac√≠o
- False si tiene un valor v√°lido

```python
# Ejemplo 1: Mostrar si cada valor es nulo o no
display(df.isna())

# Ejemplo 2: Aplicar isna() solo a la columna de edad
display(df["Age"].isna())

# Nos permite detectar si faltan algunos valores en alguna celda
```

Al combinar df-isna() con .sum() podemos contar cu√°ntos valores nulos hay por columna

```python
# Ejemplo 1: Contar los valores nulos en cada columna
display(df.isna().sum())

# Ejemplo 2: Contar solo los nulos de la columna edad
display(df["Age"].isna().sum())

# Podemos ver m√°s directamente que columnas tiene m√°s valores nulos
```

df.dropna() elimina las filas que tengan valores null

```python
# Ejemplo 1: Eliminar filas con cualquier valor nulo
display(df.dropna())

# Ejemplo 2: Eliminar solo las filas donde falte la edad
display(df.dropna(subset=["Age"]))

# Muy √∫til cuando queramos deshacernos de las columnas que tengan valores nulos
```

df.fillna() nos permite reemplazar los valores null por uno espec√≠fico o por la media, mediana, etc...

```python
# Ejemplo 1: Reemplazar valores nulos de 'Age' por la edad promedio
display(df["Age"].fillna(df["Age"].mean(), inplace=True))

# Ejemplo 2: Reemplazar todos los nulos por 0
display(df.fillna(0))

# Tambi√©n es super √∫til si no queremos borrar datos de nuestro data frame y nos sirve tener la media u otro dato en el lugar de los valores nulos
```


## üìä Step 5: Grouping and Aggregating

Functions to explore:
- df.groupby("Sex")["Survived"].mean()
- df["Fare"].mean()
- df["Age"].median()


En "df.groupby("Sex")["Survived"].mean()", groupby() agrupa todos los datos por una o varias columnas.

```python
# Ejemplo 1: Promedio de supervivencia por sexo
display(df.groupby("Sex")["Survived"].mean())

# Ejemplo 2: Promedio de supervivencia por clase y sexo
display(df.groupby(["Pclass", "Sex"])["Survived"].mean())

# Nos permite obtener datos con los que podemos hacer comparaciones de forma r√°pida
```

Con df["Fare"].mean() calculamos la media de la columna "Fare"

```python
# Ejemplo 1: Precio promedio del pasaje
display(df["Fare"].mean())

# Ejemplo 2: Precio promedio del pasaje entre los pasajeros de primera clase
display(df[df["Pclass"] == 1]["Fare"].mean())

# Nos permite ver la media y usarla para comparaciones u otras operaciones que la necesiten
```

df["Age"].median() calcula la mediana (el valor del medio) de la columna edad.

```python
# Ejemplo 1: Edad mediana de todos los pasajeros
display(df["Age"].median())

# Ejemplo 2: Clase mediana de los pasajeros
display(df["Pclass"].median())

# Nos permite conocer la mediana de manera super r√°pid
```


## üìà Step 6: Sorting and Counting

Functions to explore:
- df.sort_values("Age")
- df["Sex"].unique()
- df["Pclass"].value_counts()


df.sort_values("Age") ordena las filas de nuestro data frame seg√∫n los valores de una o varias columnas.

```python
# Ejemplo 1: Ordenar por edad de menor a mayor
display(df.sort_values("Age"))

# Ejemplo 2: Ordenar por edad de mayor a menor
display(df.sort_values("Age", ascending=False))

# Nos permite visualizar de manera ascendente o descendete las columnas que queramos, ordenadas por un valor
```

df["Sex"].unique() devuelve un array con los valores √∫nicos de la columna sexo.

```python
# Ejemplo 1: Valores √∫nicos en la columna sexo
display(df["Sex"].unique())

# Ejemplo 2: Valores √∫nicos en la columna edad
display(df["Age"].unique())

# Nos permite ver los diferentes datos que habr√° en nuestro array sin repeticiones
```

df["Pclass"].value_counts() cuenta cuantas veces aparecer√° cada valor en la columna Pclass

```python
# Ejemplo 1: Contar cu√°ntos pasajeros hay por clase
display(df["Pclass"].value_counts())

# Ejemplo 2: Mostrar el conteo ordenado de forma ascendente
display(df["Pclass"].value_counts(ascending=True))

# Nos permite ver, por ejemplo, el n√∫mero total de pasajeros en cada clase de manera r√°pida
```


## ‚öôÔ∏è Step 7: Creating or Modifying Columns

Functions to explore:
- df.assign()
- df.apply()
- df["new_col"] = ...
- pd.concat()
- pd.merge()


df.assign() permite a√±adir una o m√°s columnas al data frame sin modificar el original

```python
# Ejemplo 1: Crear una nueva columna 'MitadEdad', es la edad a la mitad
display(df.assign(MitadEdad = df["Age"] / 2).head())

# Ejemplo 2: Crear dos columnas nuevas en una sola llamada
display(df.assign(
    MitadEdad = df["Age"] / 2,
    DobleEdad= df["Age"] * 2
).head())

# √ötil por si queremos a√±adir una o m√°s columnas de manera temporal a nuestro data frame
```

df.apply() aplica una funci√≥n personalizada a una columna o fila de nuestro data frame

```python
# Ejemplo 1: Aplicar una funci√≥n a una columna (redondear el precio del pasaje)
display(df["Fare"].apply(lambda x: round(x, 1)))

# Ejemplo 2: Crear una nueva columna con apply
df["Age_Group"] = df["Age"].apply(lambda x: "Child" if x < 18 else "Adult")
display(df.head())

# nos permite crear nuevas columnas o filas cuando lo necesitemos, pudiendo aplicar diferente filtros o funciones
```

df["new_col"]= nos permite crear o modificar una columna de nuestro data frame

```python
# Ejemplo 1: Crear una nueva columna con el doble del precio
df["Fare_Double"] = df["Fare"] * 2
display(df.head(10))

# Ejemplo 2: Crear una columna indicando si el pasajero es menor de edad
df["Is_Child"] = df["Age"] < 18
display(df.head(10))

# es una forma m√°s sencilla que las anteriores de crear columnas
```

pd.concat() sirve para unir varios data frames, mediante sus filas o columnas

```python
# Ejemplo 1: Unir dos DataFrames verticalmente (a√±adir filas)
df_part1 = df.head(3)
df_part2 = df.tail(3)
display(pd.concat([df_part1, df_part2]))

# Ejemplo 2: Unir horizontalmente (a√±adir columnas)
df_a = df[["Age", "Sex"]].head(3)
df_b = df[["Fare", "Survived"]].head(3)
display(pd.concat([df_a, df_b], axis=1))

# Lo usaremos cuando queramos queramos unir diferentes partes de un dataset con otro
```

pd.merge() combina dos data frames en base a una om√°s columnas EN COM√öN.
Es parecido al JOIN de sql.

```python
# Ejemplo 1: Combinar dos DataFrames por una columna com√∫n
left = pd.DataFrame({
    "PassengerId": [1, 2, 3],
    "Age": [22, 38, 26]
})

right = pd.DataFrame({
    "PassengerId": [1, 2, 3],
    "Fare": [7.25, 71.83, 7.92]
})

display(pd.merge(left, right, on="PassengerId"))

# Ejemplo 2: Merge tipo 'inner', 'left', 'right' o 'outer'
display(pd.merge(left, right, on="PassengerId", how="outer"))

# Nos permite combinar informaci√≥n de diferentes fuentes.
```
